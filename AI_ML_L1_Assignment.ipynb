```python
import csv
import string
from nltk import sent_tokenize, word_tokenize, pos_tag, RegexpParser, ne_chunk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
with open("data_in.csv") as rfile:
    reader = csv.reader(rfile, delimiter = ' ')
    headings = next(reader)
    
    line = []
    
    for row in reader:
        for field in row:
            print()
        line.append(sent_tokenize(field))  
        
rfile.close()
```

```python
with open("data_out_sent.csv","w",newline="\n") as wfile:
    writer = csv.writer(wfile)
    writer.writerow(["Sentence Tokenization"])
    writer.writerows(line)
    
wfile.close()
```

```python
with open("data_in.csv") as rfile:
    reader = csv.reader(rfile, delimiter = ' ')
    headings = next(reader)
        
    for row in reader:
        for field in row:
            print()
        for sen in sent_tokenize(field):
            word1 = word_tokenize(sen)
            tokens_tag = pos_tag(word1)
            output = ne_chunk(tokens_tag)
            print(output)
    
rfile.close()
```

```python
f = open("NLPdataEx3&4-data_in.txt","r")

line = f.readlines()
grammer = "NP : {<DT>?<JJ>*<NN>}"
chunker = RegexpParser(grammer)

for i in line:
    cleaned_text = i.translate(str.maketrans('', '', string.punctuation))
    tag = pos_tag(word_tokenize(cleaned_text))
    output = chunker.parse(tag)
    print(tag)
    output.draw()
```

```python
f = open("NLPdataEx3&4-data_in.txt","r")

line = f.readlines()
stop_word = set(stopwords.words("english"))
porter_stemmer  = PorterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()

for w in line:
    cleaned_text = w.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(cleaned_text)
    filtered_line = [i for i in words if not i in stop_word]
    print(filtered_line)
    for word in filtered_line:
        print("Stemming for {} is {}".format(word,porter_stemmer.stem(word)))
        print("Lemma for {} is {}\n".format(word, wordnet_lemmatizer.lemmatize(word))) 
```

```python
text = open('NLPdataEx5data_senti_analyze.txt').read()
lower_case = text.lower()
cleaned_text = lower_case.translate(str.maketrans('', '', string.punctuation))

tokenized_words = word_tokenize(cleaned_text)

final_words = [i for i in tokenized_words if not i in stopwords.words('english')]

lemma_words = []
for word in final_words:
    word = WordNetLemmatizer().lemmatize(word)
    lemma_words.append(word)
    
print(lemma_words)
score=0

#created a dict file from the given NLPdataEx5dict 
with open('NLP_dictionary_file.txt', 'r') as dict_file:
    for line in dict_file:
        clear_line = line.replace("'", '').strip()
        word, marks = clear_line.split(':')
        for w in lemma_words:
            if w==word:
                score = score+int(marks)
            
    print(score)
    if score < 0:
        print("Negative Sentiment")
    elif score > 0:
        print("Positive Sentiment")
    else:
        print("Neutral Sentiment")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
filename = "auto-mpg.data"
text = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name"]
data=pd.read_csv(filename,names=text,delim_whitespace=True)
data.head()
```

```python
data = data[data.horsepower != '?']
data.horsepower = data.horsepower.astype('float')
```

```python
#independent and dependent varieable
y=data.mpg
x=data.drop('mpg', axis=1)
```

```python
#Split with 80% train and 20% test
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=15)
```

```python
#applying linear regression and training the model using train set
reg=LinearRegression()
model=reg.fit(x_train[['horsepower']],y_train)
```

```python
#coefficients coef, intercept and residues
print('Coefficients of coef : ',reg.coef_[0])
print('Coefficients of intercept : ',reg.intercept_)
```

```python
#predict using test data 
predictions=reg.predict(x_test[['horsepower']])
```

```python
#Accuracy using Mean R Square
r2=r2_score(y_test,predictions)
mse=mean_squared_error(y_test,predictions)
score=model.score(x_train[['horsepower']],y_train)
print('Score : ',score)
print('Accuracy check using R square : ',r2)
print('Mean Square Error : ',mse)
```

Multi-variate Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#function for cost/error
def rmse(Y, Y_Pred):
    rmse = np.sqrt(sum((Y-Y_Pred)**2)/len(Y))
    return rmse
```

```python
#function for gradient descent
def gradient_descent_algo(X, Y, alpha, iterations):
    n = X.shape[1]
    theta = np.zeros(n)
    m = len(X)
    j = np.dot(X, theta)
    
    cost = np.ones(iterations)
    for i in range (0, iterations):
        theta[0] = theta[0] - (alpha / m) * sum(j-Y)
        for z in range(1, n):
            theta[z]= theta[z] - (alpha/ m) * sum((j-Y) * X[:, z])
        j  = np.dot(X, theta)
        cost[i] = 1/(2*m) * sum(np.square(j-Y))  
    return cost,theta
```

```python
def graphs_mul(X,Y,cost):
    #cost vs iteration graph
    fig = plt.figure()
    plt.plot(cost)
    plt.xlabel("Iterations")
    plt.ylabel( "Cost or Loss")
    plt.show()
    
    #3D Fig Scatter
    fig2 = plt.figure()
    ax2 = plt.axes(projection='3d')
    displacement = X[:,1]
    acceleration = X[:,2]
    mpg = np.asarray(Y)
    ax2.scatter(displacement, acceleration, mpg)
    ax2.set_xlabel('Displacement')
    ax2.set_ylabel('Acceleration')
    ax2.set_zlabel('MPG')

    #3D Fig Surface
    fig3 = plt.figure()
    ax = plt.axes(projection='3d')
    displacement = X[:,1]
    acceleration = X[:,2]
    f1, f2 = np.meshgrid(displacement, acceleration)
    Z = theta[0] + theta[1] * f1 + theta[2] * f2
    ax.plot_surface(f1, f2, Z)
    ax.set_xlabel('Displacement')
    ax.set_ylabel('Acceleration')
    ax.set_zlabel('MPG')
    plt.show()
```

```python
#downloaded data
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
autompg = pd.read_csv('auto-mpg.data', sep='\s+', names = column_names)
autompg = autompg.iloc[:,:8]
autompg = autompg.apply(pd.to_numeric, errors='coerce').dropna()
```

```python
#independent and dependent values
X = autompg.iloc[:, 1:].values
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
x0 = np.ones(len(X))
X = np.column_stack((x0, X))
Y = autompg[['mpg']].values
Y = Y[:,0]
```

```python
iterations = 10000
alpha = 0.0001

cost,theta = gradient_descent_algo(X, Y, alpha, iterations)
```

```python
Y_Pred = X.dot(theta)
rmseValue = rmse(Y, Y_Pred)
rsq = r2_score(Y,Y_Pred)

print('Using Custom library')
print('RMSE = %.10f \nR Square(R2) = %.10f'%(rmseValue,rsq))
thetasym = chr(952)
print('Coefficients')
print(thetasym+'0 = %.15f'%(theta[0]))
print(thetasym+'1 = %.15f'%(theta[1]))
print(thetasym+'2 = %.15f'%(theta[2]))
print(thetasym+'3 = %.15f'%(theta[3]))
print(thetasym+'4 = %.15f'%(theta[4]))
print(thetasym+'5 = %.15f\n'%(theta[5]))
```
